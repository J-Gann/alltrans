````python
code = """
#include <iostream>
#include <fstream>
#include <string>
#include <cmath>
#include <vector>
#include <algorithm>
#include <numeric>

using namespace std;

const double NEARZERO = 1.0e-10;       // interpretation of "zero"

using vec    = vector<double>;         // vector
using matrix = vector<vec>;            // matrix (=collection of (row) vectors)

// Prototypes
void print(string title, const vec &V );
void print(string title, const matrix &A );
vec matrixTimesVector( const matrix &A, const vec &V );
vec vectorCombination( double a, const vec &U, double b, const vec &V );
double innerProduct( const vec &U, const vec &V );
double vectorNorm( const vec &V );
vec conjugateGradientSolver( const matrix &A, const vec &B );


//======================================================================


int main()
{
   matrix A = { { 4, 1 }, { 1, 3 } };
   vec B = { 1, 2 };

   vec X = conjugateGradientSolver( A, B );

   cout << "Solves AX = B\n";
   print( "\nA:", A );
   print( "\nB:", B );
   print( "\nX:", X );
   print( "\nCheck AX:", matrixTimesVector( A, X ) );
}


//======================================================================


void print( string title, const vec &V )
{
   cout << title << '\n';

   int n = V.size();
   for ( int i = 0; i < n; i++ )
   {
      double x = V[i];   if ( abs( x ) < NEARZERO ) x = 0.0;
      cout << x << '\t';
   }
   cout << '\n';
}


//======================================================================


void print( string title, const matrix &A )
{
   cout << title << '\n';

   int m = A.size(), n = A[0].size();                      // A is an m x n matrix
   for ( int i = 0; i < m; i++ )
   {
      for ( int j = 0; j < n; j++ )
      {
         double x = A[i][j];   if ( abs( x ) < NEARZERO ) x = 0.0;
         cout << x << '\t';
      }
      cout << '\n';
   }
}


//======================================================================


vec matrixTimesVector( const matrix &A, const vec &V )     // Matrix times vector
{
   int n = A.size();
   vec C( n );
   for ( int i = 0; i < n; i++ ) C[i] = innerProduct( A[i], V );
   return C;
}


//======================================================================


vec vectorCombination( double a, const vec &U, double b, const vec &V )        // Linear combination of vectors
{
   int n = U.size();
   vec W( n );
   for ( int j = 0; j < n; j++ ) W[j] = a * U[j] + b * V[j];
   return W;
}


//======================================================================


double innerProduct( const vec &U, const vec &V )          // Inner product of U and V
{
   return inner_product( U.begin(), U.end(), V.begin(), 0.0 );
}


//======================================================================


double vectorNorm( const vec &V )                          // Vector norm
{
   return sqrt( innerProduct( V, V ) );
}


//======================================================================


vec conjugateGradientSolver( const matrix &A, const vec &B )
{
   double TOLERANCE = 1.0e-10;

   int n = A.size();
   vec X( n, 0.0 );

   vec R = B;
   vec P = R;
   int k = 0;

   while ( k < n )
   {
      vec Rold = R;                                         // Store previous residual
      vec AP = matrixTimesVector( A, P );

      double alpha = innerProduct( R, R ) / max( innerProduct( P, AP ), NEARZERO );
      X = vectorCombination( 1.0, X, alpha, P );            // Next estimate of solution
      R = vectorCombination( 1.0, R, -alpha, AP );          // Residual

      if ( vectorNorm( R ) < TOLERANCE ) break;             // Convergence test

      double beta = innerProduct( R, R ) / max( innerProduct( Rold, Rold ), NEARZERO );
      P = vectorCombination( 1.0, R, beta, P );             // Next gradient
      k++;
   }

   return X;
}
"""
````

````python
class Utils:

    ################### Constants ###################

    class Models(Enum):
        CLAUDE_3_OPUS=1,
        CLAUDE_3_SONNET=2,
        CLAUDE_3_HAIKU=3,
        GPT_4o=4,
        GPT_3_TURBO=5

    class Prompts(Enum):
        system_prompt_iter = f"""
        You are an expert programmer transpiling C++ code to TypeScript code.
        You receive the source code one chunk at a time.
        Your task is to respond with a transpilation only for the chunk.
        Only respond with TypeScript code, do not add comments of your own!
        You receive the previous transpilation as help.
        Your new transpilation should continue where the last transpilation left off.
        This is the whole code, only use it as context information:
        {{code}}
        This is additional context information to help you transpile the code:
        {{context}}
        """
        user_prompt_iter = f"""
        Previous transpilation:
        {{prev_trans}}
        ======================================================
        Transpile only this chunk:
        {{chunk}}
        """

    ################### Tooling ###################

    class FileUtils:

        def _add_line_numbers(lines):
            return [f"{index + 1}: {line}" for index, line in enumerate(lines)]

        def add_line_numbers(code):
            if isinstance(code, list):
                numbered_lines = Utils.FileUtils._add_line_numbers(code)
                return numbered_lines
            else:
                lines = code.split('\n')
                numbered_lines = Utils.FileUtils._add_line_numbers(lines)
                return '\n'.join(numbered_lines)

        def read_file(file_path:str, project_path:str="", pretty=False) -> str:
            project_path = Path(project_path)
            path = project_path / file_path
            f = open(path, "r")
            text = f.read()
            f.close()
            if pretty:
                text = Utils.FileUtils.add_line_numbers(text)
            return text

        def read_lines(file_path:str, line_start, line_stop=None, project_path:str="", pretty=False):
            line_start = max(1, line_start)
            project_path = Path(project_path)
            path = project_path / file_path
            f = open(path, "r")
            lines = f.readlines()
            f.close()
            if pretty:
                lines = Utils.FileUtils.add_line_numbers(lines)
            if line_stop:
                lines = lines[line_start-1:line_stop]
            else:
                lines = lines = lines[line_start-1:]
            return lines

        def write_file(file_path:str, text: str, project_path:str=""):
            project_path = Path(project_path)
            path = project_path / file_path
            os.makedirs(os.path.dirname(path), exist_ok=True) # create missing folders
            f = open(path, "w")
            f.write(text)
            f.close()

        def write_lines(file_path:str, lines, project_path:str=""):
            project_path = Path(project_path)
            path = project_path / file_path
            f = open(path, "w")
            f.writelines(lines)
            f.close()

        def insert_lines(file_path:str, line_start, text: str, project_path:str=""):
            lines = Utils.FileUtils.read_lines(file_path, 1, None, project_path)
            lines.insert(line_start-1, text)
            Utils.FileUtils.write_lines(file_path, lines, project_path)

        def replace_lines(file_path:str, line_start, line_stop, text: str, project_path:str=""):
            lines = Utils.FileUtils.read_lines(file_path, 1, None, project_path)
            del lines[line_start-1:line_stop]
            lines.insert(line_start-1, text)
            Utils.FileUtils.write_lines(file_path, lines, project_path)

    def split_array(array, n):
        k, m = divmod(len(array), n)
        return [array[i * k + min(i, m):(i + 1) * k + min(i + 1, m)] for i in range(n)]

    def get_blocks(self, code_lines, block_size=10):
        blocks = Utils.split_array(code_lines, block_size)
        return ["\n".join(block) for block in blocks]

    def get_blocks_file(self, file_path, block_size=10):
        lines = Utils.FileUtils.read_lines(file_path, 1, None, self.target_path)
        return self._get_blocks(lines, block_size)

    def extract_code(text, language="typescript"):
        pattern = r"```"+language+"(.*?)```"
        matches = re.findall(pattern, text, re.DOTALL)
        if matches:
            return matches[0].strip()
        else:
            return text
````

````python
class Transpiler:

    # Do iterative transpilation in case the code exceeds threshold (200 lines)
    # => "chunking" by hard splitting after every 200 lines
    # => inclue last transpilation result in next prompt to continue fluently
    def transpile_iter(code_lines, llm, context="", chunk_size=200):
        chunks = Utils.split_array(code_lines, int(len(code_lines) / chunk_size)+1)
        code = "\n".join(code_lines)

        transpilations = ""
        prev_trans = ""

        messages = [ChatMessage(role="system", content=Utils.Prompts.system_prompt_iter.value.format(code=code, context=context))]

        for i, chunk in enumerate(chunks):
            chunk = "\n".join(chunk)
            messages.append(ChatMessage(role="user", content=Utils.Prompts.user_prompt_iter.value.format(prev_trans=prev_trans, chunk=chunk)))
            response = llm.chat(messages)
            transpilation = Utils.extract_code(response.message.content)
            transpilations += transpilation +"\n"
            messages.append(response.message)
            prev_trans = transpilation
        return transpilations

    def transpile_iter_file(file_path, llm, context="", chunk_size=200, project_path=""):
        code_lines = Utils.FileUtils.read_lines(file_path, 1, None, project_path)
        return Transpiler.transpile_iter(code_lines, llm, context, chunk_size)
````

````python
class TypescriptRepair:

    def __init__(self, llm, target_path, max_iterations=100):
        self.llm = llm

        self.target_path = Path(target_path)

        # Editing based on reading and writing whole files
        # Problem: Heavy on LLM context and limited by max output size of LLM
        self.EDIT_FILES_TOOLSET = [
            FunctionTool.from_defaults(fn=self.write_file),
            FunctionTool.from_defaults(fn=self.read_file),
            FunctionTool.from_defaults(fn=self.test_file),
            FunctionTool.from_defaults(fn=self.project_structure)
        ]

        self.agent = ReActAgent.from_tools(self.EDIT_FILES_TOOLSET, llm=self.llm, verbose=True, max_iterations=max_iterations)


    #################################################################

    def replace_text(self, file_path:str, text:str, replacement:str):
        """
        Replace content of file.
        file_path: Path to file
        text: Text to be replaced (at least 5 lines)
        replacement: New text
        """
        if len(text.split("\n")) < 3: return "Text not replaced! Always replace at least 5 lines at once."
        content = Utils.FileUtils.read_file(file_path, self.target_path)
        if text not in content: return "Code to be replaced not found in original document. Try again!"
        updated = content.replace(text, replacement)
        Utils.FileUtils.write_file(file_path, updated, self.target_path)

    #################################################################

    def read_selection(self)->str:
        """
        Read the selected lines
        """
        file_path, start, stop = self.selection
        return "\n".join(Utils.FileUtils.read_lines(file_path, start, stop, self.target_path, True))

    def replace_selection(self, text):
        """
        Replace the selection with new text
        text: New text to replace with
        """
        file_path, start, stop = self.selection
        Utils.FileUtils.replace_lines(file_path, start, stop, text+"\n", self.target_path)

    def insert_after_selection(self, text):
        """
        Inserts text below the selected lines
        text: Text to be inserted
        """
        file_path, start, stop = self.selection
        Utils.FileUtils.insert_lines(file_path, stop+1, text+"\n", self.target_path)

    def select_lines(self, file_path:str, start:int, stop:int)->str:
        """
        Select lines of file for further editing.
        file_path: File to select
        start: Index of first line of selection (index starting at 1)
        stop: Index of last line of selection
        """
        self.selection = (file_path, start, stop)
        return self.read_selection()

    #################################################################

    def read_lines(self, file_path:str, start:int, stop:int)->str:
        """
        Read content of a file.
        file_path: Path of file
        start: Index of first line to read (starting at index 1)
        stop: Index of last line to read
        """
        return "\n".join(Utils.FileUtils.read_lines(file_path, start, stop, self.target_path, True))

    def read_file(self, file_path:str)->str:
        """
        Read whole content of a file.
        file_path: Path of file
        """
        return Utils.FileUtils.read_file(file_path, self.target_path, False)

    def insert_lines(self, file_path:str, start:int, text:str):
        """
        Insert lines in file.
        file_path: Path of file
        start: Index of line after which text should be inserted
        text: Text to insert
        """
        Utils.FileUtils.insert_lines(file_path, start, text, self.target_path)

    def replace_lines(self, file_path:str, start:int, stop:int, text:str):
        """
        Replace lines in file.
        file_path: Path of file
        start: Index of first line to replace (starting at index 1)
        stop: Index of last line to replace
        text: Text to insert
        """
        Utils.FileUtils.replace_lines(file_path, start, stop, text, self.target_path)

    #################################################################

    def write_file(self, file_path:str, text:str):
        """
        Replace whole content of a file.
        file_path: Path of file
        text: New file content
        """
        Utils.FileUtils.write_file(file_path, text, self.target_path)

    #################################################################

    def test_file(self, file_path:str)->str:
        """Test the current code and return errors."""

        file_path = Path(self.target_path) / Path(file_path)

        compile_command = f"npx tsc --pretty {file_path}"
        compile_result = subprocess.run(compile_command, shell=True, text=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        compile_output = compile_result.stdout

        if len(compile_output) > 0: return f"""Compilation Error Report:\n-------------------------\n{compile_output}"""

        run_command = f"node {str(file_path)[:-3]}.js"   # TODO: Find cleaner way to change ts to js
        run_result = subprocess.run(run_command, shell=True, text=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        run_output = run_result.stderr

        if len(run_output) > 0: return f"""Execution Error Report:\n-----------------------\n{run_output}"""

        return f"""Execution Error Report:\n-----------------------\nNo errors found!"""

    def project_structure(self)->str:
        """Show all folders and files in the source folder"""
        startpath = str(self.target_path)
        res = ""
        for root, dirs, files in os.walk(startpath):
            level = root.replace(startpath, '').count(os.sep)
            indent = ' ' * 4 * (level)
            res += f'{indent}{os.path.basename(root)}/' + "\n"
            subindent = ' ' * 4 * (level + 1)
            for f in files:
                res += f'{subindent}{f}' + "\n"

        return "\n".join(res.split("\n")[1:])

    #################################################################

    def run(self):
        agent_response = self.agent.chat("""Your task is to fix errors in TypeScript files. Start by looking at the project structure and test the code. Finish when the test results in no errors.""")
````

````python
class AllTransCpp:
    def __init__(self, llm_openai, project_path, target_path):
        self.llm_openai = llm_openai
        self.project_path = project_path
        self.target_path = target_path
        self.signatures = self.get_project_signatures()


    # TODO: Use aider for signature extraction?
    # TODO: Do not generate description for every module? => save compute!
    def get_file_signatures(self, path):
        class Signature(BaseModel):
            """
            One component of the code like a function or a class.
            signature: Name of the component with parameters and types
            description: Description of the implementation of the component
            """
            signature: str
            description: str

        class Signatures(BaseModel):
            """
            All signatures in the code
            """
            signatures: List[Signature]

        prompt_template_str = "Extract all relevant components of the following code: {code}"
        program = OpenAIPydanticProgram.from_defaults(output_cls=Signatures, prompt_template_str=prompt_template_str, verbose=False, llm=self.llm_openai)

        code = Utils.FileUtils.read_file(path)

        signatures = program(code=code).signatures

        signatures_res = []

        for sig in signatures:
            signatures_res.append({
                "path": path,
                "signature": sig.signature,
                "description": sig.description
            })


        return signatures_res

    def get_project_signatures(self, as_string=True, file_ending=".cpp"):
        signatures = []
        source_files = glob.glob(os.path.join(self.project_path, '**', f'*{file_ending}'), recursive=True)
        for path in source_files:
            sig = self.get_file_signatures(path)
            signatures.extend(sig)
        if as_string:
            return "PROJECT SIGNATURES\n-----------------------------\n"+"\n-------------------------\n".join([f"""
path: {r['path']}
signature: {r['signature']}
description: {r['description']}
""" for r in signatures])


        return signatures


    def get_includes_cpp(self, code):
        """
        Gets all imports of the cpp code
        """
        parser = get_parser('cpp')
        tree = parser.parse(code.encode())

        def traverse_tree(node):
            imports = []
            if node.type == 'preproc_include':
                imports.append(node.text.decode("utf-8"))
            for child in node.children:
                imports.extend(traverse_tree(child))
            return imports

        return traverse_tree(tree.root_node)

    def get_dependency_mappings(self, path):

        class ModuleResolverTypescript(BaseModel):
            """
            Mapping of one component of the source language to a equivalent component of the target language.
            Your mapping should help to move codebase from one language to the other.

            source_module_name: Name of source module
            source_module_description: Description of functionality of source module

            target_module_type: Type of the component in the target language (dependency, built-in or local)

            target_module_name: Name of target module
            target_module_description: Short description of functionality of target module
            target_module_example: Example of how to use the module in the code
            """

            source_module_name: str
            source_module_description: str

            target_module_type: str

            target_module_name: str
            target_module_description: str
            target_module_example: str

        class ModuleResolverListTypescript(BaseModel):
            """
            List of modules
            """
            module_list: List[ModuleResolverTypescript]


        code = Utils.FileUtils.read_file(path, self.project_path)


        system_message = ChatMessage(role="system", content="You are a programming assistant helping to find suitable dependencies in other languages based on given code.")
        user_message = ChatMessage(role="user", content=f"""Based on the following C++ code, which TypeScript dependenies are equivalent to the C++ dependencies?
        For every dependency in the C++ code, explicitly mention an alternative by either stating module names built-in into the TypeScript language or external dependencies.
        Mention which dependencies are local dependencies.
        For each elternative dependency, give an example on how to use it in the given code to achieve the same functionality: {code}""")
        messages = [system_message, user_message]
        context_dependencies = llm.chat(messages)



        prompt_template_str = """
Here is the complete code for reference {code}.
Here is context information about the source code of the current project: {context_signatures}.
Here is context information about alternative dependencies: {context_dependencies}
Map the following modules from {source_language} to {target_language}: {modules}"""
        program = OpenAIPydanticProgram.from_defaults(output_cls=ModuleResolverListTypescript, prompt_template_str=prompt_template_str, verbose=False, llm=llm_openai)
        modules = self.get_includes_cpp(code)
        mapping = program(source_language="C++", target_language="TypeScript", modules="".join(modules), context_signatures=self.signatures, code=code, context_dependencies=context_dependencies)

        return f"""
Dependency mapping information from C++ to TypeScript:
------------------------------------------------------
""" + "\n\n".join([f"""
source module name: {map.source_module_name}
source module description: {map.source_module_description}
target module type: {map.target_module_type}
target module name: {map.target_module_name}
target module description: {map.target_module_description}
target module example: {map.target_module_example}
""" for map in mapping.module_list])


    def transpile_file_with_project_context(self, path):
        dependency_mappings = self.get_dependency_mappings(path)
        context = self.signatures + "\n\n" + dependency_mappings
        print(context)
        transpilation = Transpiler.transpile_iter_file(path, llm, context, 200, self.project_path)
        return transpilation


    def transpile_project(self, file_ending=".cpp"):
        # TODO: How to approach this?
        # - random?
        # - dependency graph and start at leaves?
        # - leave to ReAct agent?
        source_files = glob.glob(os.path.join(self.project_path, '**', f'*{file_ending}'), recursive=True)
        for file_path in source_files:
            transpilation = self.transpile_file_with_project_context(file_path)
            Utils.FileUtils.write_file(Path())
            print(file_path)

trans = AllTransCpp(llm_openai, "C-Snake-Game", "Transpilation")
transpilation = trans.transpile_file_with_project_context("main.cpp")
#trans.transpile_project()
````
